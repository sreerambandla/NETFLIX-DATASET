```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Dataset

```python
df=pd.read_csv(r"C:\Users\bandl\Downloads\Netflix Dataset.csv")
df
```

```python
sns.heatmap(df.isnull()) 
```

# 1. Handle Missing Values

```python
# ---- Missing Values Visualization ----
# 1. Heatmap
plt.figure(figsize=(15,5))
sns.heatmap(df.isnull(), cbar=False, cmap="PiYG")
plt.title("Missing Values Heatmap", fontsize=10)
plt.show()

# 2. Bar Plot of Missing Values
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)

plt.figure(figsize=(10,4))
sns.barplot(x=missing_data.index, y=missing_data.values, palette="magma")
plt.title("Missing Values Count by Column", fontsize=10)
plt.ylabel("Count of Missing Values")
plt.xticks(rotation=45)
plt.show()
```

### What this will show:
- Heatmap → Highlights where missing values.
- Bar plot → Shows exact missing values per column.

# 2 Remove Duplicates

```python
# ---- 2. Bar Plot of Missing Values ----
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)

plt.figure(figsize=(6,4))
sns.barplot(x=missing_data.index, y=missing_data.values, palette="magma")
plt.title("Missing Values Count by Column", fontsize=10)
plt.ylabel("Count of Missing Values")
plt.xticks(rotation=45)
plt.show()
```

### What this does:
- Prints how many duplicates exist before cleaning.
- Removes duplicates using drop_duplicates().
- Confirms duplicates removed (should show 0 after).
- Shows dataset shape after cleaning.

# 5 Non-Visual Bivariate Analysis

```python
# ---- 1. Categorical vs Categorical (Category vs Rating) ----
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Rating", hue="Category", palette="Set2")
plt.title("Category vs Rating", fontsize=15)
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.show()

# ---- 2. Categorical vs Numerical (Duration by Category) ----
if 'Duration_num' in df.columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x="Category", y="Duration_num", data=df, palette="coolwarm")
    plt.title("Duration by Category", fontsize=15)
    plt.ylabel("Duration (Minutes/Seasons)")
    plt.show()

# ---- 3. Numerical vs Numerical (Correlation Heatmap) ----
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
plt.figure(figsize=(6,4))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="YlOrBr", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=15)
plt.show()

# ---- 4. Release Year vs Category (Trend) ----
plt.figure(figsize=(12,6))
sns.countplot(data=df, x="Release Year", hue="Category", palette="pastel")
plt.title("Release Year vs Category", fontsize=15)
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.show()
```

### Observations:

* 1. Category vs Rating
- Movies spread across multiple ratings like PG, R, TV-MA.
- TV Shows mostly fall in TV-MA, TV-14, TV-PG (more adult-oriented).
* 2. Duration by Category
- Movies have higher duration variance (many around 90–120 minutes).
- TV Shows’ duration measured in seasons, usually 1–3 seasons.
* 3. Correlation Heatmap
- Very weak correlation between Release Year and Duration.
- No strong linear relationships, meaning each variable is mostly independent.
* 4. Release Year vs Category
- A huge spike in releases after 2015, mostly TV Shows.

# Final Summary 

## 1. Issues Found in the Raw Dataset
- Missing values in key columns like Director, Cast, Country, and Date_added.
- Duplicates present in rows, which could mislead statistics.
- Inconsistencies in categorical data (e.g., "Movie" vs "movie", "TV Show" vs "tv show", extra spaces).
### Incorrect data types:
- Date_added stored as string instead of datetime.
- Release Year not numeric.
- Duration was mixed text (e.g., "90 min", "2 Seasons").

## 2. How Issues Were Resolved
### Missing Values:
- Filled categorical nulls with "Unknown" or mode.
- Filled dates with NaT (treated as missing).
### Duplicates:
-Removed using df.drop_duplicates().
### Inconsistencies:
- Converted text columns to lowercase.
- Removed extra spaces and standardized categories.
### Data Types:
- Converted Date_added → datetime.
- Converted Release Year → integer.
- Split Duration into Duration_num (numeric) and Duration_type (minutes/seasons).
- Converted categorical features (Category, Rating, Country) to category type.


## 3. Summary of Bivariate Relationships
- Category vs Rating:
- Movies have diverse ratings (PG, R, TV-MA).
- TV Shows mainly rated TV-MA, TV-14, TV-PG.
### Category vs Duration:
- Movies average ~100 minutes, with large variance.
- TV Shows measured in seasons (commonly 1–3).
### Release Year Trends:
- Strong rise in content after 2015, especially TV Shows (Netflix Originals).
### Numerical Correlations:
- Weak correlation between release year and duration → newer content does not strongly affect runtime.
